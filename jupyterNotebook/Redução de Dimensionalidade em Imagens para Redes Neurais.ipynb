```python
import tkinter as tk
from tkinter import filedialog
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def rgba_to_rgb(path):
    img = mpimg.imread(path)
    if img.shape[2] == 4:
        img = img[:, :, :3]
    return img
```

```python
def converter_para_cinza(img):
    pesos = np.array([0.299, 0.587, 0.114])
    img_cinza = np.dot(img, pesos).astype(np.uint8)
    return img_cinza
```

```python
def binarizar(img):
    threshold = int(np.mean(img))
    img_bin = np.where(img >= threshold, 255, 0).astype(np.uint8)
    return img_bin
```

```python
path = './dataset/Dog/1.jpg'

img = rgba_to_rgb(path)

cinza  =  converter_para_cinza(img)


bina = binarizar(cinza)

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

axs[0].imshow( img)
axs[0].set_title("Original")
axs[0].axis('off')

axs[1].imshow(cinza)
axs[1].set_title("Cinza")
axs[1].axis('off')

axs[2].imshow(bina, cmap='gray')
axs[2].set_title("Bina")
axs[2].axis('off')

plt.tight_layout()
plt.show()
```
